```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
from tabulate import tabulate
```

```python
dataset1 = pd.read_csv('./train.csv', encoding='utf-8')
dataset1
```

```python
print(dataset1.dtypes.to_string())
```

```python
df1_int = dataset1.select_dtypes(include="int")
df1_int.dtypes
```

```python
df1_numeric = dataset1.select_dtypes([np.number])
df1_numeric.dtypes
```

```python
df1_numeric.describe()
```

```python
df1_numeric['SalePrice'].describe()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12,5))

sns.boxplot(data=df1_numeric, y="SalePrice", ax=axes[0])
sns.histplot(data=df1_numeric, x="SalePrice", ax=axes[1])
sns.kdeplot(data=df1_numeric, x="SalePrice", ax=axes[2])

fig.suptitle("Sale Price Distribution")
axes[0].set_title("Box Plot")
axes[1].set_title("Histogram")
axes[2].set_title("KDE")
plt.subplots_adjust(wspace=0.3)
plt.show()
```

```python
df1_numeric['LotArea'].describe()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12,5))

sns.boxplot(data=df1_numeric, y="LotArea", ax=axes[0])
sns.histplot(data=df1_numeric, x="LotArea", ax=axes[1])
sns.kdeplot(data=df1_numeric, x="LotArea", ax=axes[2])

fig.suptitle("Lot Area Distribution")
axes[0].set_title("Box Plot")
axes[1].set_title("Histogram")
axes[2].set_title("KDE")
plt.subplots_adjust(wspace=0.3)
plt.show()
```

```python
sns.jointplot(data=df1_numeric, x="LotArea", y="SalePrice")
plt.show()
```
