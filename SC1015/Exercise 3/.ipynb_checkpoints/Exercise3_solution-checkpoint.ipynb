```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.core import display as ICD
from math import ceil
from scipy import stats
```

```python
df_main = pd.read_csv("./train.csv")
```

### Problem 1 : Analysis of Numeric Variables

```python
cols_a = ['LotArea', 'GrLivArea', 'TotalBsmtSF', 'GarageArea']
df_main[cols_a].isnull().any()
```

a) Which of these variables has the maximum number of outliers as per box-plot? How many outliers does it have?

```python
fig, axes = plt.subplots(ceil(len(cols_a)/2), 2, figsize=(10,10))
fig.suptitle("Boxplots", fontsize=18, ha='center')
plt.subplots_adjust(wspace=0.3, hspace=0.3)

most_outliers = ([], 0)

for i, col in enumerate(cols_a):
    df_col = df_main[col]
    q1, q3 = df_col.quantile([0.25, 0.75])
    iqr = q3 - q1
    outliers = df_col[(df_col < q1-1.5*iqr) | (df_col > q3+1.5*iqr)]

    num_outliers = outliers.size
    if most_outliers[1] < num_outliers: most_outliers = ([col], num_outliers)
    elif most_outliers[1] == num_outliers: most_outliers[0] = most_outliers[0].append(col)
    print(f"{col} has {num_outliers} outliers: {outliers.values.tolist()}")
        
    ax = axes[i//2, i%2]
    sns.boxplot(data=df_main, y=col, ax=ax)
    ax.set_title(f"Boxplot of {col} ({num_outliers} outliers)")
    ax.set_xticks([])
    
plt.show()
```

```python
print(f"Most outliers ({most_outliers[1]}): {', '.join(most_outliers[0])}")
```

### Problem 2 : Analysis of Categorical Variables

```python
cols_b = ['MSSubClass', 'Neighborhood', 'BldgType', 'OverallQual']
df_main[cols_b].isnull().any()
```

a) Which of these variables has the highest number of levels? Which of the levels has the highest number of houses?

```python
most_levels = ([], 0)

for i, col in enumerate(cols_b):
    num_levels = len(df_main[col].unique())
    if most_levels[1] < num_levels: most_levels = ([col], num_levels)
    elif most_levels[1] == num_levels: most_levels[0] = most_outliers[0].append(col)
    print(f"Number of levels in {col}: {num_levels}")
    
    df_col_count = df_main.groupby(by=col).size().sort_values(ascending=False).reset_index()
    df_col_count.columns = [col, 'num_elem']
    display(df_col_count)
    df_top_level = df_col_count.head(1)
    print(
        f"Class '{df_top_level[col].values[0]}' in column '{col}'",
        f"has the most number of houses ({df_top_level['num_elem'].values[0]})\n\n\n"
    )
```

```python
print(f"Most levels ({most_levels[1]}): {', '.join(most_levels[0])}")
```

b) Choose the top two variables that you think will help us the most in predicting ‘SalePrice’ of houses in this data.

```python
fig, axes = plt.subplots(ceil(len(cols_b)/2), 2, figsize=(10,10))
plt.subplots_adjust(wspace=0.5, hspace=0.3)

for i, col in enumerate(cols_b):
    df_corr_b = df_main['SalePrice'].to_frame().join(df_main[col].to_frame())
    col_order = (df_corr_b
                 .groupby(by=col)
                 .mean()
                 .sort_values(by='SalePrice', ascending=False)
                 .reset_index()[col]
                 .astype(str)
                 .values
                 .tolist()
    )
    df_corr_b[col] = df_corr_b[col].astype(str)
    
    if not col == 'SalePrice':
        ax = axes[i//2, i%2]
        sns.boxplot(data=df_corr_b, x='SalePrice', y=col, order=col_order, ax=ax)
        
    ax.set_title(f"{col} vs SalePrice")
    
plt.show()
```

Top 2 variables would be OverallQual and Neighbourhood, as they seem to have the strongest correlation

### Problem 3 : Interesting Questions for EDA

a) Does the SalePrice of a house get affected by whether it has a Garage or not? Justify your answer using EDA.
<br><br>
<i>Conclusion:</i> There is a positive correlation between having a garage and sale price. However, the sample size on houses without garages is very limited.
<br><br>
<i>Note:</i> Positive correlation between garage area and sale price. This may be due to larger (and thus more expensive) properties having more room to accomodate larger garages while still maintaining similar proprtion of residence to garage area.

```python
cols_garage_categorical = [
    'GarageType', 'GarageYrBlt', 'GarageFinish',
    'GarageQual', 'GarageCond', 'GarageCars', 'MiscFeature'
]
cols_garage_numeric = ['GarageArea']

df_garage = df_main[['SalePrice', *cols_garage_categorical, *cols_garage_numeric]].copy()
df_garage = df_garage.rename(columns={'MiscFeature': 'Garage2nd'})
df_garage['Garage2nd'] = np.where(df_garage['Garage2nd'] == 'Gar2', 'Yes', 'No')
cols_garage_categorical.append('Garage2nd')
cols_garage_categorical.remove('MiscFeature')

df_garage['GarageYrBlt'] = (df_garage['GarageYrBlt']//10)*10
df_garage['GarageYrBlt'] = df_garage['GarageYrBlt'].fillna(0).astype(int).astype(str).replace('0', np.nan)
df_garage = df_garage.rename(columns={'GarageYrBlt': 'GarageDecadeBlt'})
cols_garage_categorical.append('GarageDecadeBlt')
cols_garage_categorical.remove('GarageYrBlt')

df_garage['HasGarage'] = np.where(df_garage['GarageType'].isnull(), 'No', 'Yes')
```

```python
fig, axes = plt.subplots(
    ceil((len(cols_garage_categorical)+len(cols_garage_numeric))/2), 2, figsize=(10,15)
)
plt.subplots_adjust(wspace=0.5, hspace=0.5)

for i, col in enumerate([*cols_garage_categorical, *cols_garage_numeric]):
    if col == 'SalePrice': continue

    df_garage_group = df_garage['SalePrice'].to_frame().join(df_garage[col].to_frame())
    col_order = (df_garage_group
                 .groupby(by=col)
                 .mean()
                 .sort_values(by='SalePrice', ascending=False)
                 .reset_index()[col]
                 .astype(str)
                 .values
                 .tolist()
    )

    ax = axes[i//2, i%2]
    ax.set_title(f"{col} vs SalePrice")
    if col in cols_garage_categorical:
        df_garage_group[col] = df_garage_group[col].astype(str)
        sns.boxplot(data=df_garage_group, x='SalePrice', y=col, order=col_order, ax=ax)
    elif col in cols_garage_numeric:
        df_garage_group = df_garage_group.join(df_garage['HasGarage'])
        df_garage_group = df_garage_group[df_garage_group['HasGarage'] == 'Yes']
        sns.regplot(data=df_garage_group, x='SalePrice', y=col, ax=ax, scatter_kws={'s':1})

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,5))
plt.subplots_adjust(wspace=0.5)
sns.boxplot(data=df_garage, x='SalePrice', y='HasGarage', ax=axes[0])
sns.stripplot(data=df_garage, x='SalePrice', y='HasGarage', ax=axes[1], s=3)

fig.suptitle("Garage vs Sales Price")
axes[0].set_title("Boxplot")
axes[1].set_title("Stripplot")

for idx, ax in enumerate(axes):
    xticks = ax.get_xticks().astype(int)
    xticks = xticks[xticks > 0]
    ax.set_xticks(xticks)
    ax.set_xticklabels(xticks, rotation=337.5, rotation_mode='anchor', ha='left')
    
plt.show()
```

b) Does the SalePrice of a house get affected by how recently it got Remodeled? Justify your answer using EDA.
<br><br>
<i>Conclusion:</i> Positive relationship between recency of remodel date and sale price.

```python
cols_remodel_numeric = ['YearRemodAdd', 'YearBuilt']
cols_remodel_categorical = []
df_remodel = df_main[cols_remodel_numeric].copy()

for idx, col in enumerate(cols_remodel_numeric):
    new_name = col.replace('Year', 'Decade')
    df_remodel[new_name] = (df_remodel[col].astype(int)//10)*10
    df_remodel[new_name] = df_remodel[new_name].astype(str)
    cols_remodel_categorical.append(new_name)

df_remodel = df_remodel.join(df_main['SalePrice'].to_frame())
cols_remodel_numeric.append('SalePrice')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,5))
plt.subplots_adjust(wspace=0.5)

for i, col in enumerate(cols_remodel_categorical):
    if col in cols_remodel_numeric: continue
    col_order = (df_remodel[[col, 'SalePrice']]
             .groupby(by=col)
             .mean()
             .sort_values(by='SalePrice', ascending=False)
             .reset_index()[col]
             .astype(str)
             .values
             .tolist()
    )
    
    axes[i].set_title(f"{col} vs SalePrice")
    sns.boxplot(data=df_remodel, x='SalePrice', y=col, order=col_order, ax=axes[i])

plt.show()
```
